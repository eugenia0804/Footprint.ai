```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
#pytrends = TrendReq(hl='en-US', tz=360, timeout=(10,25), proxies=['https://172.241.112.79:80',], retries=2, backoff_factor=0.1, requests_args={'verify':False})
```

```python
kw_list = ["Northwestern", "Football"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
#pytrends.interest_over_time()
pytrends.build_payload(kw_list=['pizza', 'bagel'], timeframe=['2022-09-04 2022-09-10', '2022-09-18 2022-09-24'])
pytrends.multirange_interest_over_time()
```
